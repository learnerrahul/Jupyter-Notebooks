## Filtering out Best and Worst movies - profit/revenue/ROI


### Integrating HTML for better viewing experience

```python
from IPython.display import HTML
pd.options.display.max_colwidth = 200
```

### Best Movies

```python
movies_best = movies[["poster_path", "title", "budget_musd", "revenue_musd",
              "vote_count", "vote_average", "popularity"]].copy()
movies_best
```

```python
movies_best["profit_musd"] = movies.revenue_musd.sub(movies.budget_musd)
movies_best["return"] = movies.revenue_musd.div(movies.budget_musd)
movies_best
```

```python
movies_best.columns = ["", "Title", "Budget", "Revenue", "Votes", 
                   "Average Rating", "Popularity", "Profit", "ROI"]
```

```python
movies_best.set_index("Title", inplace = True)
movies_best
```

```python
movies_best.iloc[1,0]
```

### Improving Asthetics of the data by experimenting on a subset


```python
subset = movies_best.iloc[:10, :2]
subset
```

```python
HTML(subset.to_html(escape=False))
```

# Evauluating performances- Best Movies


```python
movies_best.Budget.fillna(0, inplace = True)
movies_best.Votes.fillna(0, inplace = True)
movies.info()
```

```python
movies_best.loc[movies_best.Budget >= 5].sort_values(by = "ROI", ascending = False)
```

```python
#Creating fuction  best_worst for ease of excecution with changeable parameters

def best_worst(n, by, ascending = False, min_bud = 0, min_votes = 0):
    
    
    movies2 = movies_best.loc[(movies_best.Budget >= min_bud) & (movies_best.Votes >= min_votes), 
                      ["", by]].sort_values(by = by, ascending = ascending).head(n).copy()
    
    return HTML(movies2.to_html(escape=False))
```

## Performance Analysis - Best and Worst


### Top 10 movies - Highest Budget

```python
best_worst(10, "Budget")
```

### Top 10 movies - Highest Renue

```python
best_worst(10,"Revenue")
```

### Top 10 movies - Most Profitable

```python
best_worst(10,'Profit')
```

### Top 10 movies -  Least Profitable

```python
best_worst(10,'Profit', ascending= True)
```

### Top 10 movies -  Highest ROI

```python
best_worst(10,'ROI',min_bud=20)
```

### Top 10 movies -  Lowest ROI

```python
best_worst(10,'ROI',min_bud=20,ascending = True)
```

### Top 10 movies -  Movies with most number of votes

```python
best_worst(10, 'Votes',min_votes=100)
```

### Top 10 movies -  Highest Rated Movies

```python
best_worst(10,'Average Rating',min_votes=100)
```

### Top 10 movies -  Highest Rated Movies - Most Popular movies

```python
best_worst(10,'Popularity',min_votes=100)
```

## Franchises Vs Stand-Alone Movies - Which one More Successful?

```python
movies.belongs_to_collection
```

```python
movies["Franchise"] = movies.belongs_to_collection.notna()
```

```python
movies.Franchise
```

```python
movies.Franchise.value_counts()
```

###  Franchise Vs Stand-alone: Average Revenue 

```python
movies.groupby("Franchise").revenue_musd.mean()
```

###  Franchise Vs Stand-alone : Budget

```python
movies.groupby("Franchise").budget_musd.mean()
```

###  Franchise Vs Stand-alone :ROI and Profitability

```python
movies["ROI"] = movies.revenue_musd.div(movies.budget_musd)
```

```python
movies.groupby("Franchise").ROI.median()
```

###  Franchise Vs Stand-alone : Popularity

```python
movies.groupby("Franchise").popularity.mean()
```

###  Franchise Vs Stand-alone : Average Rating

```python
movies.groupby("Franchise").vote_average.mean()
```

```python
movies.groupby("Franchise").agg({"budget_musd": "mean", "revenue_musd": "mean", "vote_average": "mean",
                            "popularity": "mean", "ROI":"median", "vote_count":"mean"})
```

### Conclusion - Franchises tend to be more profitable than Stand-alone projects
